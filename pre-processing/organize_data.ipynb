```python
import pandas as pd
import csv
import numpy as np
import os
import glob
import shutil
```

```python
my_df = pd.DataFrame()
for i in range(13):
    print("gdc_sample_sheet.2020-05-20-"+str(i+1)+".tsv")
    df = pd.read_csv("gdc_sample_sheet.2020-05-20-"+str(i+1)+".tsv", index_col=None, header=0, sep='\t')
    my_df = pd.concat([my_df, df], axis=0)
```

```python
my_df
```

```python
my_df = my_df[['File ID','File Name','Data Category','Project ID','Case ID']]
my_df
```

```python
names = ["TCGA-BRCA","TCGA-GBM","TCGA-OV","TCGA-LUAD","TCGA-UCEC","TCGA-KIRC","TCGA-HNSC","TCGA-LGG","TCGA-THCA","TCGA-LUSC","TCGA-PRAD","TCGA-SKCM","TCGA-COAD","TCGA-STAD","TCGA-BLCA","TCGA-LIHC","TCGA-CESC","TCGA-KIRP","TCGA-SARC","TCGA-LAML","TCGA-PAAD","TCGA-ESCA","TCGA-PCPG","TCGA-READ","TCGA-TGCT","TCGA-THYM","TCGA-KICH","TCGA-ACC","TCGA-MESO","TCGA-UVM","TCGA-DLBC","TCGA-UCS","TCGA-CHOL"]
dirFile = 'cancer_types/'
dirFile1 = 'data/data/'
```

```python
for i in range(len(names)):
    temp = my_df.loc[my_df['Project ID'] == names[i]]
    for index, row in temp.iterrows():
        if row['Data Category'] == 'Clinical':
            shutil.move(dirFile1+row['File Name'], dirFile+names[i]+"/clinical/"+row['File Name'])
        elif row['Data Category'] == 'Transcriptome Profiling':
            if "counts" in row['File Name']:
                shutil.move(dirFile1+row['File Name'], dirFile+names[i]+"/exp_count/"+row['File Name'])
            elif "FPKM-UQ" in row['File Name']:
                shutil.move(dirFile1+row['File Name'], dirFile+names[i]+"/exp_upper/"+row['File Name'])
            elif "FPKM." in row['File Name']:
                shutil.move(dirFile1+row['File Name'], dirFile+names[i]+"/exp_norm/"+row['File Name'])
        elif row['Data Category'] == 'DNA Methylation':
            shutil.move(dirFile1+row['File Name'], dirFile+names[i]+"/myth/"+row['File Name'])
        elif row['Data Category'] == 'Copy Number Variation':
            shutil.move(dirFile1+row['File Name'], dirFile+names[i]+"/cnv/"+row['File Name'])
        elif row['Data Category'] == 'Simple Nucleotide Variation':
            shutil.move(dirFile1+row['File Name'], dirFile+names[i]+"/vcf/"+row['File Name'])
```
