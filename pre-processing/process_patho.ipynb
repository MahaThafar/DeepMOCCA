```python
import pandas as pd
import csv
import numpy as np
import os
import glob
```

```python
dirFile = 'cancer_types/TCGA-BRCA/vcf/output'
search_path = dirFile + "/*.txt.dat"
allFile = glob.glob(search_path)
```

```python
for file_ in allFile:
    print(file_)
    df = pd.read_csv(file_, index_col=None, header=None, sep=' ')
    df.columns=['gene','value']
    newdf = df[df['value'] != '.']
    newdf1 = newdf[newdf['gene'] != '.']
    newdf1.to_csv(file_, sep='\t', index=False, header=False)
```
